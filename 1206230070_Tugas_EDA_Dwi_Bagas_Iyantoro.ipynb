```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Histogram of a numerical variable
plt.hist(data['lunch'])
plt.xlabel('lunch')
plt.ylabel('test preparation course')
plt.title('test preparation distribution')
plt.show()
```

```python
# Bar plot for a categorical variable
sns.countplot(data['gender'])
plt.title('Gender Distribution')
plt.show()
```

```python
numerik = data[['math score', 'reading score', 'writing score']]
corr_matrix = numerik.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title('Indikator Korelasi')
plt.show()
```
